```python
input_countries = data_con['Country']

countries = {}
for country in pycountry.countries:
    countries[country.name] = country.alpha_3
countries
```

```python
codes = [pycountry.countries.get(name = country).alpha_3 for country in input_countries]
```

```python
data_con['Code'] = codes
data_con.head()
```

```python
# df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_world_gdp_with_codes.csv')
scl = [[1.0, 'rgb(84,39,143)'], [0.8, 'rgb(117,107,177)'],[0.6, 'rgb(158,154,200)'],[0.4, 'rgb(188,189,220)'],\
       [0.2, 'rgb(218,218,235)'], [0.0, 'rgb(242,240,247)']]


df = data_con
data = [ dict(
        type = 'choropleth',
        locations = df['Code'],
        z = df['count'],
        text = df['Country'],
        colorscale = 'Viridis',
        autocolorscale = False,
        reversescale = True,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            autotick = False,
            tickprefix = '$',
            title = 'No of Crashes'),
      ) ]

layout = dict(
    title = 'Aviation Accidents by Country',
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'Mercator'
        )
    )
)

fig = dict( data=data, layout=layout )
py.iplot( fig, validate=False, filename='d3-world-map' )
```
